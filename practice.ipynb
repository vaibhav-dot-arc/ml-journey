```python
#DAY 2

# CALUCULATOR USING FUNCTIONS

def add(a,b):
  return a+b

def subtract(a,b):
  return a-b

def multiply(a,b):
  return a*b

def division(a,b):
  if b == 0 :
    return "undefined"
  else :
    return a/b



def calculator(num1,operator,num2):
  if operator == "+" :
    return add(num1,num2)
  elif operator == "-" :
    return subtract(num1,num2)
  elif operator == "*":
    return multiply(num1,num2)
  elif operator == "/" :
    return division(num1,num2)
  else :
    return "invalid operator"


print(calculator(1, "+" ,2 ))
print(calculator(1, "-" ,2 ))
print(calculator(1, "*" ,2 ))
print(calculator(1, "/" ,2 ))
print(calculator(10, "/", 0))   # should print: undefined
print(calculator(10, "$", 5))   # should print: invalid operator
```

```python

# ARGS AND KWARGS

#1 default

def greet(msg, name="ravi"):
  print(f"{name},{msg}")


greet("good morning",)


#2 ARGS

def total(*marks):
  result =  0

  for a in marks :
   result += a

  return result

print(total(80,90,10,20))


#3 kwargs


def info(**details):


  for key,value in details.items() :
     print(f"{key} : {value}")

info(name="vaibhav", age=28, city="hyderabad")
```

```python
#DAY 3

#######DICTIONARIES##########

phonebook = {}

phonebook["ravi"]="1234567890"
phonebook["ram"]="0987654321"
phonebook['noname']="6969696969"

print("All contacts")
for key, value in phonebook.items():
  print(f"{key}'s phonenumber is {value}")



search = "ram"
if search in phonebook :
  print(f"{search}'s phonenumber is {phonebook[search]}")
else :
  print(f"{search} does not exist")

del phonebook["noname"]
print(f"\nAfter deleting noname: {phonebook}")

print(f"total contacts are {len(phonebook)}")

```
